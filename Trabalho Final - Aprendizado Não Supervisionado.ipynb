## Clusterização das características - k-means

# Determinando o melhor k:

```python
sum_squared_dist = []
K = range(1,50)
for k in K:
    km = KMeans(n_clusters=k, random_state=0)
    km = km.fit(features)
    sum_squared_dist.append(km.inertia_)

plt.plot(K, sum_squared_dist, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('O método Elbow mostrando o melhor k')
plt.show()
```

## Clusterizando com o melhor k

```python
data_array = features.values
kmeans = KMeans(n_clusters=6, init='k-means++', n_init=10, random_state=42)
features["cluster"] = kmeans.fit_predict(data_array)
features.groupby("cluster").aggregate("mean").plot.bar(figsize=(10,7.5))
plt.title("Características clusterizadas - k-means")
plt.show()
```

## Dispersão dos grupos no espaço

```python
fig = plt.figure(figsize = (15,15))
ax = fig.add_subplot(111, projection='3d')
ax.set_xlabel('RM', fontsize = 15)
ax.set_ylabel('LSTAT', fontsize = 15)
ax.set_zlabel('PTRATIO', fontsize = 15)
ax.set_title('Dispersão 3D entre as características - k-means', fontsize = 20)

targets = [0, 1, 2, 3, 4, 5]
colors = ['r', 'g', 'b', 'c', 'm', 'y']
for target, color in zip(targets,colors):
    indicesToKeep = features['cluster'] == target
    ax.scatter(features.loc[indicesToKeep, 'RM'],
               features.loc[indicesToKeep, 'LSTAT'],
               features.loc[indicesToKeep, 'PTRATIO'],
               c = color,
               s = 50)
ax.legend(targets)
ax.grid()
```

## Clusterização das características - GMM

```python
data_array = features.values
gmm = GaussianMixture(n_components=6)
gmm.fit(data_array)
features["cluster"] = gmm.predict(data_array)
features.groupby("cluster").aggregate("mean").plot.bar(figsize=(10,7.5))
plt.title("Características clusterizadas - GMM")
plt.show()
```

```python
fig = plt.figure(figsize = (15,15))
ax = fig.add_subplot(111, projection='3d')
ax.set_xlabel('RM', fontsize = 15)
ax.set_ylabel('LSTAT', fontsize = 15)
ax.set_zlabel('PTRATIO', fontsize = 15)
ax.set_title('Dispersão 3D entre as características - GMM', fontsize = 20)

targets = [0, 1, 2, 3, 4, 5]
colors = ['r', 'g', 'b', 'c', 'm', 'y']
for target, color in zip(targets,colors):
    indicesToKeep = features['cluster'] == target
    ax.scatter(features.loc[indicesToKeep, 'RM'],
               features.loc[indicesToKeep, 'LSTAT'],
               features.loc[indicesToKeep, 'PTRATIO'],
               c = color,
               s = 50)
ax.legend(targets)
ax.grid()
```
